```python
# Plotting Financial charts using library mplfinance (mplf)
# By default, it shows OHLC chart.

Reliance_2021 = pd.read_json('Reliance_2021.json')
mplf.plot(Reliance_2021)
```

```python
# Plotting Candlestick charts using library mplfinance (mplf)

mplf.plot(Reliance_2021,type='candle')
```

```python
# Plotting Renko charts using library mplfinance (mplf)

mplf.plot(Reliance_2021,type='renko')
```

```python
# Plotting PNF charts using library mplfinance (mplf)

mplf.plot(Reliance_2021,type='pnf')
```

```python
# Plotting the Candlestick chart using library mplfinance with Volume and moving average(mplf)

mplf.plot(Reliance_2021,type='candle',volume=True,mav=(5,10,20))
```

```python
# Saving the Candlestick chart using library mplfinance with Volume and moving average(mplf)

mplf.plot(Reliance_2021,type='candle',volume=True,mav=(5,10,20),savefig='Candlestickchart.png')
print("Completed")
```

```python
# Plotting the same as above without using mplfinance lib (only using pandas and matplotlib)
# Moving averages can be calculated and plotted along with Volume. But OHLC data cannot be plotted as candlesticks.

#Setting the sub plots:

ax1 = plt.subplot2grid((6,1),(0,0),rowspan=4, colspan=1) # 6 rows and 1 column total - This sub plot is at 0,0 - occupies 4 rows size.

ax2 = plt.subplot2grid((6,1),(4,0),rowspan=2, colspan=1, sharex=ax1) # 6 rows and 1 column total - This sub plot is at 4,0 - occupies 2 rows size.

ax1.xaxis_date()

#Deriving Moving average:

Reliance_2021['10dma'] = Reliance_2021['Adj Close'].rolling(window=10,min_periods=0).mean()

ax1.plot(Reliance_2021.index, Reliance_2021['Adj Close'])
ax1.plot(Reliance_2021.index, Reliance_2021['10dma'])
ax2.fill_between(Reliance_2021.index, Reliance_2021['Volume'])

plt.tight_layout()
plt.show()
```

```python
# Correlation of stock prices

combodf =  pd.DataFrame()

combodf['Reliance']= Reliance ['Adj Close']
combodf['MangalamCem']= MangalamCem ['Adj Close']
combodf['SBI']= SBI ['Adj Close']

print ('Printing share prices together:')
print (combodf)

print ('Printing Correlation:')
correlation = combodf.corr()
print (correlation)

print ('Plotting Correlation:')

plt.matshow(correlation)
plt.show()
```
