```python
from pandas_datareader import data as web
import datetime as dt

start = dt.datetime(2017,1,1)
end = dt.datetime(2021,5,24)
df = web.DataReader('RELIANCE.NS','yahoo',start,end)

# Loaded stock prices are exported to a CSV file.
df.to_csv('Reliance.csv',sep=";")
print ('Done')
```

```python
import matplotlib.pyplot as plt
df['Adj Close'].plot()
plt.show
```

```python
# Importing styles and use labels.
from matplotlib import style
style.use ('ggplot')
df['Adj Close'].plot()
plt.ylabel ('Adjucated Close')
plt.title ('Reliance Share price')
plt.show
```

```python
# Plotting two stocks to compare.
Reliance = web.DataReader('RELIANCE.NS','yahoo',start,end)
TCS = web.DataReader('TCS.NS','yahoo',start,end)
Reliance['Adj Close'].plot(label='Reliance')
TCS['Adj Close'].plot(label='TCS')
plt.ylabel('Adjucated Close')
plt.title ('Share price')
plt.legend (loc='upper left')
plt.show()
```
